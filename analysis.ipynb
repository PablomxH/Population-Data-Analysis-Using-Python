# Comparative density/growth rate

```python
#DataFrame with values of density per each continent
#first, define df
df_result = pd.DataFrame(columns=['Continent','Density','Growth Rate'])

#exctract every continent in a list
names=df_continents['Continent'].values.tolist()

#a function for filter a density of every country in a continent
def sum_density(parameter):
    #filter for continent
    filter_continent=population[population['Continent']==parameter]
    
    #sumatory of every value
    density=filter_continent.loc[:,['Density (per km²)']].sum()
    grow=filter_continent.loc[:,['Growth Rate']].sum()
    
    #df of density
    df_density=pd.DataFrame({'Continent':parameter,'Density':density.values,'Growth Rate':grow.values})
    return df_density

#in this loop df of density it's append in the final df
#for continent in names:
    #df_result = df_result.append(sum_density(continent), ignore_index=True)
    #df_result = pd.concat([df_result, sum_density(continent)], ignore_index=True)

df_result = pd.concat(df_continents['Continent'].apply(sum_density).tolist(), ignore_index=True)
    
#print(df_result.to_string(index=True))
print('Continents by Growth Rate')
display(df_result.sort_values(by='Growth Rate', ascending=False))

print('Continents by Density')
display(df_result.sort_values(by='Density', ascending=False))

#sumatory of square kilometers in the earth
print('Earth: '+ str(population.loc[:,['Area (km²)']].sum()))
```

```python
#Graphs
plt.style.use('dark_background')

cont= df_result['Continent']
den= df_result['Density']
gro=df_result['Growth Rate']

fig, axes =plt.subplots(1,2,figsize=(10,5))

axes[0].bar(cont,den,width=0.4,color=['yellow','#00FFFF'])
axes[1].bar(cont,gro,width=0.4,color=['yellow','#00FFFF'])

axes[0].tick_params(axis='x', rotation=45)
axes[1].tick_params(axis='x', rotation=45)

axes[0].set_xlabel('Continents')
axes[0].set_ylabel('per km²')

axes[1].set_xlabel('Continents')
axes[1].set_ylabel('Growth Rate')

axes[0].set_title('Density by Continent')
axes[1].set_title('Growth Rate by Continent')

plt.show()
```
